```
#1.Which ones were the most used city for the three days?
n.columns
```

```
a=n.groupby(['VILLES']).sum()
a.sort_values(by='v',ascending=False,inplace=True)
a
```

```
#Which cities were the most used during business and home hours?
#splitting the columns
datetime = n
datetime = datetime['d'].apply(lambda x: pd.Series([i for i in (x.lower().split(' '))]))
datetime.columns = ['d','t']
datetime = datetime['t'].apply(lambda x: pd.Series([i for i in (x.lower().split(':'))]))
datetime.columns = ['h','m','s']
datetime['h']= pd.to_numeric(datetime['h'])
datetime.dtypes
```

```
win = pd.merge(n, datetime, left_index=True , right_index=True)
win
```

```
o_hrs = win[(win['h'] > 8) & (win['h'] < 17)]
o_hrs
```

```
h_hrs = win[(win['h'] < 8)]
h_hrs2 = win[(win['h'] >= 17)]
h_hrs = pd.concat([h_hrs,h_hrs2] , ignore_index = 0)
h_hrs2.drop
h_hrs
```
