#4.Data Cleaning

4.1 Removing white spaces between columns 

```
#This is to improve validity of dataframe
df0 = df.rename(columns={'Bluecar counter':'Bluecar_counter', 'Utilib counter':'Utilib_counter', 'Utilib 1.4 counter':'Utilib_1.4_counter', 
                               'Charge Slots':'Charge_Slots', 'Charging Status':'Charging_Status', 'Displayed comment':'Displayed_comment', 
                               'Geo point':'Geo_point', 'Postal code':'Postal_code', 'Public name':'Public_name', 'Rental status':'Rental_status', 
                               'Scheduled at':'Scheduled_at', 'Station type':'Station_type',
                          'Subscription status':'Subscription_status'})
df0.head(3)
```

4.2 Dropping unneccessary rows


```
#Drop rows where cars are equal to zer0 to improve Accuracy
df0 = df0.drop(df0[df0.Cars == 0].index)
df0.shape
```

4.3 Checking for Null values and filling them 

```
#This action is to check for completeness of data
df0.isna().sum()
```

```
#Filling the null values with zero
df1 = df0.fillna(0)
df1.head(3)
```

4.4 Checking for Duplicates

```
#This is to checks if the data is consistent. The results shows it is.
df1.duplicated().sum()
```

4.5 Renaming and changing data type for the Columns

```
#this is to ensure uniformity
#changing data types
df2 = df1.astype({"Cars": float, "Bluecar_counter": float, 'Utilib_counter': float, 'Utilib_1.4_counter': float, 'Charge_Slots':float})
df2.info()
```

```
#renaming columns
df2.rename(columns = {'Bluecar_counter':'bcc','Utilib_counter':'utc','Utilib_1.4_counter':'utc1.4','Charging_Status':'cs','Rental_status':'rs', 'Postal_code':'pc','Station_type':'st'}, inplace = True)
```

#5.Analysis
Research Question

Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.

```


#Joining the cars of a station by group by

df2['Usage'] = df2['utc1.4'] + df2['utc1.4'].astype(float)
df2.head()
#calculation for the pickecd cars
df2['Picked'] = df2['Cars'] - df2['Usage'].astype(float)
#calculatig for the returned cars 
df2['Returned'] = df2['Cars'] + df2['Usage'].astype(float)
df2.head()
```

```
#selecting the paris city
df3 = df2.loc[df2['City'] == 'Paris']

# selecting for the picking up hours
dfx = df3.groupby(['hour'])['Picked'].max()
dfy = dfx.sort_values(ascending=False)
dfy.head(1)
```
